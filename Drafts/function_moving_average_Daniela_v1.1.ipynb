```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
df=pd.read_csv("synchronous machine.csv",decimal=',',delimiter=';')
```

```python
def moving_average (filter_length):
    Iy_moving_average = np.convolve(df['Iy'], np.ones((filter_length)), mode ="same")
    Iy_moving_average /= filter_length
    Iy_moving_average= Iy_moving_average[round(filter_length/2): round(len(df['Iy'])-filter_length/2)]
    original_function= df['Iy']
    original_Iy= original_function[round(filter_length/2): round(len(df['Iy'])-filter_length/2)] 
    
    PF_moving_average = np.convolve(df['PF'], np.ones((filter_length)), mode ="same")
    PF_moving_average /= filter_length
    PF_moving_average= PF_moving_average[round(filter_length/2): round(len(df['PF'])-filter_length/2)]
    original_function= df['PF']
    original_PF= original_function[round(filter_length/2): round(len(df['PF'])-filter_length/2)] 
    
    e_moving_average = np.convolve(df['e'], np.ones((filter_length)), mode ="same")
    e_moving_average /= filter_length
    e_moving_average= e_moving_average[round(filter_length/2): round(len(df['e'])-filter_length/2)]
    original_function= df['e']
    original_e= original_function[round(filter_length/2): round(len(df['e'])-filter_length/2)] 
    
    dIf_moving_average = np.convolve(df['dIf'], np.ones((filter_length)), mode ="same")
    dIf_moving_average /= filter_length
    dIf_moving_average= dIf_moving_average[round(filter_length/2): round(len(df['dIf'])-filter_length/2)]
    original_function= df['dIf']
    original_dIf= original_function[round(filter_length/2): round(len(df['dIf'])-filter_length/2)]    
    
    If_moving_average = np.convolve(df['If'], np.ones((filter_length)), mode ="same")
    If_moving_average /= filter_length
    If_moving_average= If_moving_average[round(filter_length/2): round(len(df['If'])-filter_length/2)]
    original_function= df['If']
    original_Iy= original_function[round(filter_length/2): round(len(df['If'])-filter_length/2)] 
    
    concatenated_dataframes= np.stack((Iy_moving_average, PF_moving_average, e_moving_average, dIf_moving_average, If_moving_average ), axis=1)
    df_filtered_moving_average = pd.DataFrame(concatenated_dataframes, columns=['Iy', 'PF', 'e', 'dIF', 'If'])
    
    
    return df_filtered_moving_average 
```
